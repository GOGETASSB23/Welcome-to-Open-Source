MATPLOTLIB

USEFUL FOR MAKING PLOTS

```python
import numpy as np #to get data
import pandas as pd # to create a dataframe
import matplotlib.pyplot as plt #to make plots
```

```python
x = np.linspace(0, 10, 100)#this creates 100 space separated values between 0 to 10
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
print(y)
print(z)
```

```python
#sin wave
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
#cosine wave
plt.show()
```

```python
#adding title x axis and ya axis labels
plt.plot(x,y)
plt.xlabel('angle') #writes angle in x axis
plt.ylabel('sine value')
plt.title('sine wave')
#gives title to our plot

```

```python
#parabola
x = np.linspace(-10, 10, 20)
y = x**2;
plt.plot(x,y)

plt.title('upward parabola')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x, y, 'rx')
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-' )#g- represents green color and with - sign
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

BAR PLOTS

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1]) #0,0 represents origin and i,1 rep ht and width
languages = ['ENGLISH', 'SPANISH', 'GERMAN', 'FRENCH', 'HINDI']
people = [100, 50, 60, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])
languages = ['ENGLISH', 'SPANISH', 'GERMAN', 'FRENCH', 'HINDI']
people = [100, 50, 60, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

SACTTER PLOT


```python
X = np.linspace(20, 40, 30)
Y = np.sin(X)
Z = np.cos(X)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(X,Y,color='g')
ax.scatter(X,Z,color='b')
plt.show()
```

3D SCATTER PLOT

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```

SEABORN
USED FOR DATA VISUALISATION

```python
#importing the library
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

NOTE: SEABORN HAS SOME BUILT IN DATA SETS



```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

```python
#visualise the data
sns.relplot(data=tips, x = 'total_bill', y = 'tip' , col = 'time', hue = 'smoker', style='smoker', size = 'size')
```

```python
#load the iris data  --it is about petal length and other details about 3 types of iris flowers
iris = sns.load_dataset('iris')

```

```python
iris.head()
```

scattr plot

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length', hue='species',data = iris )
```
